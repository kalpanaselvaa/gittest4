## Problem 1: The Score Standardizer

**The Scenario:**
You are a teacher. You have exams from different years, but they were graded differently! 
- Year 2020 exams were out of 10 points.
- Year 2021 exams were out of 100 points.
- Year 2022 exams might be out of 50 points, or 200 points - who knows!

You need a way to compare them fairly. You want to convert *all* scores to a standard scale between **0.0 and 1.0**.

**The Logic:**
To standardize a score, you need to know the `minimum` and `maximum` scores from that batch.
- The lowest score in the batch becomes **0.0**.
- The highest score in the batch becomes **1.0**.
- Everything else falls proportionally in between.

The formula is:
$$ \text{standardized\_score} = \frac{\text{score} - \text{min}}{\text{max} - \text{min}} $$

**Your Task:**

1. `analyze_scores(scores)`: Look at a list of scores and "remember" the min and max.
2. `standardize(scores)`: Take a list of scores and return a new list of standardized (0.0-1.0) scores using the remembered min/max.
